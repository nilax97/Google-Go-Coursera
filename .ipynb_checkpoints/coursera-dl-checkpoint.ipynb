```python
import os

homedir = os.getcwd()

for f in files:
    file_dir = "/".join(f.split("/")[:-1])
    file_name = f.split("/")[-1]
    os.chdir(file_dir)
    if os.path.exists(file_name):
        print("EXISTS " + f)
        os.chdir(homedir)
        continue
    os.system('cat ' + file_name + '.* >' + file_name)
    os.system("md5 " + file_name + " > " + file_name + "2md5")
    md51 = open(file_name + "md5").read()
    md52 = open(file_name + "2md5").read()
    if md51 != md52:
        print("MD5 ERROR",file_name)
        os.system("rm " + file_name)
    os.chdir(homedir)
```

```python
import os
import sys
import ast

main_dir = ['google-golang-specialization/']
count = 0
total = 0

def gitup(path, file_count):
    global count
    global total
    if os.path.isfile(path):
        size = os.path.getsize(path)
        total += size
        print(count,"/",file_count,path)
#         os.system('find . -name ".DS_Store" -delete')
        os.system('git add ' + "'" + path + "'")
        if((total//(1024*1024))>500):
            os.system('git commit -m "updates"')
            os.system('git push -u origin master')
            total = 0
        count += 1
    else:
        for file in sorted(os.listdir(path)):
            gitup(os.path.join(path,file),file_count)

for x in main_dir:
    count = 0
    file_count = sum(len(files) for _, _, files in os.walk(x))
    gitup(x,file_count)
```
